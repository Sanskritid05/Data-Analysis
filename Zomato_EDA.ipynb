```python
ratings = final.groupby(['Aggregate rating' , 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:"Rating Count"})
ratings
```

```python
total_ratings = ratings['Rating Count'].sum()
# green
total_good = ratings.loc[ratings['Rating text'].isin(['Excellent' , 'Very Good']) , 'Rating Count'].sum()
# average
total_average = ratings.loc[ratings['Rating text'].isin(['Good' , 'Average']) , 'Rating Count'].sum()
# poor
total_poor = ratings.loc[ratings['Rating text'] == 'Poor' , 'Rating Count'].sum()
# zero 
total_zeros = ratings.loc[ratings['Rating text'] == 'Not rated' , 'Rating Count'].sum()

print('Rating Nutshell\n')
print('Total Users :' , total_ratings + total_zeros)
print("Total Ratings : ",total_ratings)
print("Good Ratings : ",total_good)
print("Average Ratings : ",total_average)
print("Poor Ratings : ",total_poor)
print("Zero Ratings : ",total_zeros)

```

### Rating Nutshell Keypoints
1. Total Users : 11699          (total_ratings + total_zeros)
2. Good Ratings :  1380         (4.0 to 4.9)
3. Average Ratings :  5837      (2.5 to 3.9)
4. Poor Ratings :  186          (1.8 to 2.4)
5. Zero Ratings :  2148         (0.0 to 0.0)

```python
cuisine_key = final.Cuisines.value_counts()[:10].keys()
cuisine_values = final.Cuisines.value_counts()[:10].values
top_10_cuisines = pd.DataFrame({'Cuisines':cuisine_key , 'Count': cuisine_values})
top_10_cuisines
```

```python
### figure of 2 columns (1 for table , 1 for chart)

fig , ax = plt.subplots(1, 2 , figsize=(14,5))

# Left side show table
ax[0].axis('tight')
ax[0].axis('off')

table = ax[0].table(cellText=top_10_cuisines.values,
                   colLabels=top_10_cuisines.columns, 
                   loc='center',
                   cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)

# Right side show graph
ax[1].barh(top_10_cuisines['Cuisines'] , top_10_cuisines['Count'] , color = 'orange')
ax[1].set_xlabel('Count')
ax[1].set_title('Top 10 Cuisines')
ax[1].invert_yaxis()

plt.tight_layout()
plt.show()
```
